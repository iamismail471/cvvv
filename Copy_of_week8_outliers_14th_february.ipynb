```python
#Techniques for outlier detection and removal:
#Z-score treatment :
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/nethajinirmal13/Training-datasets/main/diabetes.csv')
df.sample(5)
```

```python
sns.distplot(df["Glucose"])
sns.distplot(df['Age'])
```

```python

import warnings
warnings.filterwarnings('ignore')
plt.subplot(1,2,1)
sns.distplot(df["Insulin"])
plt.subplot(1,2,2)
sns.distplot(df['Outcome'])
```

```python
#Finding the Boundary Values
print("Highest allowed",df['SkinThickness'].mean() + 3*df['BloodPressure'].std())
print("Lowest allowed",df['BloodPressure'].mean() - 3*df['BloodPressure'].std())
```

```python

#Finding the Outliers
df[(df['Insulin'] > 127.172890) | (df['Age'] < 11.038047)]

```

```python
#Trimming of Outliers
new_df = df[(df['BloodPressure'] < 127.172890) & (df['BloodPressure'] > 11.038047)]
new_df
```
